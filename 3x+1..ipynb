```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def  collatz ( num ): 
    l = []
    while (num!= 1 ): 
       if (num% 2 == 0 ): 
          num=num// 2 
          l += [[len(l) , num]]
       else : 
          num= 3 *num+ 1 
          l += [[len(l), num]]
    return l
```

```python
for n in range(2,1000):
    cn = pd.DataFrame(collatz(n))
    plt.plot(cn[:][0].tolist(), cn[:][1].tolist())
print(max(collatz(n)))
```
