# 회귀
## 카테고리 분류: 1,2,3
## 회귀: 같은 분류의 특정 수치 찾기
### 최근접이웃 회귀: 가까운 애들의 수치의 평균을 가져온다

```python
import pandas as pd
```

```python
df = pd.read_csv('data/fish.csv')
```

```python
df.info()
```

```python
perch = df[df['Species'] =='Perch']
```

```python
import matplotlib.pyplot as plt

plt.scatter(perch['Weight'], perch['Length2'])
```

```python
from sklearn.model_selection import train_test_split
tri,tei,trt,tet = train_test_split(perch[['Length2']], perch[['Weight']])
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knr = KNeighborsRegressor()
```

```python
knr.fit(tri, trt)
```

```python
knr.score(tei,tet)
```
